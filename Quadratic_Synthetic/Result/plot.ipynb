```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn_avg = pd.read_csv('FedAvg_-23.0_None_None_None_None_None_2022.csv', usecols=[1], header = 0, names=['fn_avg']) # alpha1 = -23
fn_avg = fn_avg['fn_avg'].to_list()
```

```python
fn_giant = pd.read_csv('GIANT_-34.0_None_None_None_None_None_2022.csv', usecols=[1], header = 0, names=['fn_giant']) # alpha1 = -34
fn_giant = fn_giant['fn_giant'].to_list()
```

```python
fn_disco = pd.read_csv('DiSCO_None_None_None_None_None_None_2022.csv', usecols=[1], header = 0, names=['fn_disco'])
fn_disco = fn_disco['fn_disco'].to_list()
```

```python
fn_disco_prac = pd.read_csv('DiSCO_-1.0_None_None_None_None_None_2022.csv', usecols=[1], header = 0, names=['fn_disco_prac']) # alpha1 = -1
fn_disco_prac = fn_disco_prac['fn_disco_prac'].to_list()
```

```python
fn_fedh_n = pd.read_csv('FedHybrid_0.0_0.0_0.0_-5.0_14.0_10_2022.csv', usecols=[1], header = 0, names=['fn_fedh_n']) # FedH-N
fn_fedh_n = fn_fedh_n['fn_fedh_n'].to_list()
```

```python
fn_fedh_g = pd.read_csv('FedHybrid_-23.0_5.0_0.0_0.0_17.0_0_2022.csv', usecols=[1], header = 0, names=['fn_fedh_g']) # FedH-G
fn_fedh_g = fn_fedh_g['fn_fedh_g'].to_list()
```

```python
fn_fedh_8 = pd.read_csv('FedHybrid_-17.0_12.0_0.0_-5.0_14.0_8_2022.csv', usecols=[1], header = 0, names=['fn_fedh_8']) # the two gradient ones are one high cond and one normal client, 3, 9
fn_fedh_8 = fn_fedh_8['fn_fedh_8'].to_list()
```

```python
fn_fedh_8_new = pd.read_csv('FedHybrid_-17.0_12.0_0.0_-5.0_14.0_8.csv', usecols=[1], header = 0, names=['fn_fedh_8']) # the two gradient ones are two normal clients, 3, 5
fn_fedh_8_new = fn_fedh_8_new['fn_fedh_8'].to_list()
```

```python
fn_fedh_5_all_cond = pd.read_csv('FedHybrid_-17.0_10.0_0.0_-4.0_15.0_5_2022_[8, 4, 7, 9, 1].csv', usecols=[1], header = 0, names=['fn_fedh_5_all_cond']) # the five Newton clients include all high cond clients
fn_fedh_5_all_cond = fn_fedh_5_all_cond['fn_fedh_5_all_cond'].to_list()
```

```python
fn_fedh_5 = pd.read_csv('FedHybrid_-16.0_-4.0_0.0_-5.0_15.0_5_2022_[8, 3, 7, 5, 1].csv', usecols=[1], header = 0, names=['fn_fedh_5']) # the five Newton clients include two high cond clients, clients_cond=[8, 4, 7, 9]
fn_fedh_5 = fn_fedh_5['fn_fedh_5'].to_list() 
```

```python
fn_pn_dg = pd.read_csv('PN_DG_None_3.0_0.0_None_4.0_None_2022_None.csv', usecols=[1], header = 0, names=['fn_pn_dg'])
fn_pn_dg = fn_pn_dg['fn_pn_dg'].to_list()
```

```python
plt.plot(np.log(fn_avg[:2000]), 'k')
plt.plot(np.log(fn_giant[:2000]), 'y')
plt.plot(np.log(fn_disco[:2000]), 'g:')
plt.plot(np.log(fn_disco_prac), 'g')
plt.plot(np.log(fn_fedh_n), 'r')
plt.plot(np.log(fn_fedh_g[:2000]), 'k:')
plt.plot(np.log(fn_fedh_8), 'm')
#plt.plot(np.log(fn_fedh_8_new), 'm:')
#plt.plot(np.log(fn_fedh_5), 'b')
plt.plot(np.log(fn_fedh_5_all_cond), 'b')
plt.plot(np.log(fn_pn_dg), 'y:')

#ax = plt.subplot(111)
legend = ['FedAvg', 'GIANT', 'DiSCO', 'DiSCO_prac', 'FedH-N', 'FedH-G', 'FedH-8', 'FedH-5', 'PN-DG']
#ax.legend(legend, bbox_to_anchor=(1.1, 1.05))
plt.legend(legend, loc='lower right')

plt.rcParams.update({'font.size': 12})
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

plt.xlabel('Communication Round, '  + r'$r$')
plt.ylabel('Log Optimality Gap, ' + r'$\log(f(x_0^r) - f^*)$')

filename = 'Quadratic_Synthetic_S1_f.pdf'
plt.savefig(fname = filename)
```
